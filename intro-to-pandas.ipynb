```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV into DataFrame
df = pd.read_csv("demo.csv")

# Display first few rows
print("ðŸ“„ First 5 rows:")
print(df.head())

# Summary statistics
print("\nðŸ“Š Summary Statistics:")
print(df.describe())
```

```python
# Bar chart: Average salary per department
avg_salary = df.groupby('Department')['Salary'].mean()
avg_salary.plot(kind='bar', title='Average Salary by Department', color='skyblue')
plt.ylabel('Salary')
plt.xlabel('Department')
plt.tight_layout()
plt.show()
```

```python
# Line plot: Salary vs Age
df.plot(x='Age', y='Salary', kind='line', marker='o', title='Salary vs Age')
plt.ylabel('Salary')
plt.xlabel('Age')
plt.tight_layout()
plt.show()
```

```python
# Pie chart for employee count by department
dept_counts = df['Department'].value_counts()
dept_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, title='Employees per Department', ylabel='')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Age vs Salary
df.plot(kind='scatter', x='Age', y='Salary', color='green', title='Age vs Salary')
plt.grid(True)
plt.tight_layout()
plt.show()
```
