```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
marks = list(np.random.randint(1, 100, (10)).astype(int))
names = ['Brinda', 'Mahek', 'Jiya', 'Dhruvi', 'Marva', 'Hetal', 'Priya', 'Gunjan', 'Priyanshi', 'Gayatri']
marks, names
```

```python
df = pd.DataFrame(columns=['Marks', 'Name'])
```

```python
df['Marks'] = marks
df['Name'] = names
df
```

```python
plt.pie(df['Marks'], labels=df['Name'])
```

```python
plt.bar(df['Name'], df['Marks'])
```

```python
plt.plot(df['Name'], df['Marks'])
```

```python
plt.hist(df['Marks'])
```

```python
plt.scatter(df['Marks'], df['Name'])
```
